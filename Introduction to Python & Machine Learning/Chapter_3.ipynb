```python
strr="Either I win or Die!Win"
print(strr)
print(len(strr))

pstrr=strr.lower()
print(pstrr)

# whether a substr is present or not
print("either" in pstrr)
print("win" in pstrr)

#index from the starting
Sindx=pstrr.find("win")
print(Sindx)
#index from the end
Eindx=pstrr.rfind("win")
print(Eindx)

#how many times its present
ct=pstrr.count("win")
print(ct)

#replace a substr
strr2=strr.replace("win","winn")
print(strr2);
```

## Splitting, joining, formatted string

```python
prompt="It's me vs me"
prompt2="My-ego-haha"
token=prompt.split()
token2=prompt2.split("-")
print(token)
print(token2)

sentence="-".join(token)
print(sentence)
```
