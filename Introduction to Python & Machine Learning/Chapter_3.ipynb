```python
strr="Either I win or Die!Win"
print(strr)
print(len(strr))

pstrr=strr.lower()
print(pstrr)

# whether a substr is present or not
print("either" in pstrr)
print("win" in pstrr)

#index from the starting
Sindx=pstrr.find("win")
print(Sindx)
#index from the end
Eindx=pstrr.rfind("win")
print(Eindx)

#how many times its present
ct=pstrr.count("win")
print(ct)

#replace a substr
strr2=strr.replace("win","winn")
print(strr2);
```

## Splitting, joining, formatted string

# List

```python
## declaration 
numbers = [100,20,30,40,50,100,100] # int 
print(numbers)
float_numbers = [10.20,3.15,2.4,2.7,3.5] # float 
print(float_numbers)
fruits = ['apple', 'orange' , 'lichi'] # string 
print(fruits)
mix = [10 , 10.20 , 'apple'] # mixed list
print(mix)

l_list =[ [100,20,30] , [14,12,31] , [13,12,12] ]
print(l_list)

print(l_list[0])
print(l_list[0][2])
```

```python
## access 

print(numbers[2])

numbers[2] = 90

print(numbers[2])

new_list = numbers[2:5]

print(new_list)
```

```python
## modification 

## adding a item 
numbers.append(888)
print(numbers)

#insert at any pos
numbers.insert(2,1111)
print(numbers)
## deleting the last value 
numbers.pop()

print(numbers)

## deleting a specific values 
numbers.remove(100)

print(numbers)
```

```python
numbers.sort()
print(numbers)
```

```python
numbers.sort(reverse=True) #numbers.reverse()
print(numbers)
```

## List as Stack

```python
stack=[]

stack.append(1)
stack.append(2)
stack.append(4)
stack.append(5)
# print(stack)

print(f"top element: {stack[-1]}")
stack.pop()
print(f"top element: {stack[-1]}")
print(f"size of the st is: {len(stack)}")
print(f"the stack is : {stack[::-1]}")
```

## List as Queue

```python
queue = []

# insertion
queue.append(1)
queue.append(2)
queue.append(3)
queue.append(4)
queue.append(5)
print(queue)

print(queue.pop(0))
print(queue)
print(f"front e {queue[0]}")
```
