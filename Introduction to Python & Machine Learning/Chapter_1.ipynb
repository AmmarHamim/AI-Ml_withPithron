## 3.Operators in Python

##### 3.1 Arithmetic Operators

```python
x = 10 ; y = 3 

sum = x + y # summation 
sub = x - y # subtraction
mult = x * y # multiplication
div = round(x / y , 2) # division 

rem = x % y # remainder 

print(sum , sub , mult, div , rem)
```

##### 3.2 Comparison (Relational) Operators

```python
x = 10 ; y = 3 

greater_than = x > y 

greater_than_equal = (10 >= 11)

print(greater_than , greater_than_equal)


less_than = x < y 

less_than_equal = (10 <= 10)

print(less_than , less_than_equal)



equal =  x==y

print(equal)


not_equal = x != y

print(not_equal)
```

##### 3.3 Logical Operators

```python
x=10;y=20;z=30
result=(z>y) and (y>x)
print(result)
```

```python
x = True
y = False

print(x and y)  # False
print(x or y)   # True
print(not x)    # False
```

##### 3.3.1 Precedence and Associavity

```python
eq = 10 + 10/2 - 5*2
print(eq)

eq = (10+10)/2 - 5*2 

print(eq)
```

##### 4.2 Max Min

```python
a,b,c=map(int,input().split())
m=max(a,b,c)
mm=min(a,b,c)

print(mm, m)
```

##### 4.3 Even, Odd, Positive and Negative

```python
# n=int(input())
num=list(map(int,input().split()))
n=len(num)

pos=neg=even=odd=0
for x in num:
    if(x>0):
        pos+=1
    if(x<0):
        neg+=1
    if(x%2==0):
        even+=1
    if(x%2==1):
        odd+=1

print("Even: ",even)
print("Odd: ",odd)
print("Positive: ",pos)
print("Negative: ",neg)
```
