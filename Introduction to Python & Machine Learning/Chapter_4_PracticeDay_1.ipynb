## 1. Smart Light Pilot

```python
brightness,threshold=map(float,input().split())

if brightness>=threshold:
    print("ON")
else:
    print("OFF")
```

## 2. Sensor Calibrator — Min-Max Scaling

```python
x, min_v, max_v=map(float,input().split())
norm = (x - min_v) / (max_v - min_v)

print(f"{norm:.2f}")
```

## 3. Mini Loss Tracker

```python
n=int(input())
tar=float(input())
losses=[float(input()) for _ in range(n)]

# avg=sum(losses)/n

total=0;
for x in losses:
    total+=x
avg=total/n
if avg<=tar:
    print("PASS")
else:
    print("RETRY")
```

## 4. Majority Vote Gate

```python
n=int(input())
votes=[input() for _ in range(n)]
# print(votes)
ct_yes=ct_no=0
for x in votes:
    if x=="YES":
        ct_yes+=1;
    else:
        ct_no+=1

if ct_yes>=ct_no:
    print("ACCEPT")
else:
    print("REJECT")
```

## 5. AI Mood Detector

```python
msg=input()
if "happy" in msg or "joy" in msg or "smile" in msg:
    print("Happy Mood")
elif "sad" in msg or "angry" in msg or "cry" in msg:
    print("Sad Mood")
else:
    print("Neutral Mood")
```

## 6. AI Brightness Analyzer

```python
values=list(map(int,input().split()))
# print(values)
# avg=sum(values)/len(values)
total=0
for x in values:
    total+=x
avg=total/len(values)
# print(total)

if avg>170:
    print("Bright Image")
elif avg>=85:
    print("Normal Image")
else:
    print("Dark Image")
```

## 7. AI Bias Detector

```python
pred=list(input().split())
# print(pred)

ct_a=pred.count("A")

percentage_a=ct_a/len(pred) * 100
percentage_a=round(percentage_a,2)
# print(percentage_a)
if percentage_a>70 or percentage_a < 30:
    print("Biased Model")
else:
    print("Fair Model")
```

## 8. AI Keyword Counter

```python
keywords = ["ai", "data", "model", "learn", "train", "neural"]
message = input().split()
ct=0
for word in message:
    if word in keywords:
        ct+=1
# ct = sum(1 for word in message if word in keywords)
if ct >= 2:
    print("AI Detected")
else:
    print("Not AI Related")
```
