## Problem Solving

##### Max Min

```python
a,b,c=map(int,input().split())
m=max(a,b,c)
mm=min(a,b,c)

print(mm, m)
```

##### Even, Odd, Positive and Negative

```python
# n=int(input())
num=list(map(int,input().split()))
n=len(num)

pos=neg=even=odd=0
for x in num:
    if(x>0):
        pos+=1
    if(x<0):
        neg+=1
    if(x%2==0):
        even+=1
    if(x%2==1):
        odd+=1

print("Even: ",even)
print("Odd: ",odd)
print("Positive: ",pos)
print("Negative: ",neg)
```

##### Digits

```python
t=int(input())

for i in range(t):
    num= int(input())
    if num==0:
        print(0)
        continue
    while num>0:
        print(num%10,end=" ")
        num//=10
    print()
```
