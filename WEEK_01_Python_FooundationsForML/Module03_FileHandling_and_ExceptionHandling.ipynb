## Practice 1 :

Get the text from sample.txt , and count the following


Number of lines

Number of words

Number of characters

and save it to the counter_of_string.txt file

```python
with open("./sample_data/sample.txt","r") as f:

  str_lst=f.readlines()
  #num of lines
  print(str_lst)
  print(f"num of lines = {len(str_lst)}")

  #num of words
  words  = list(map(lambda x: len(x.split()),str_lst))
  print(words)

  totalWords=sum(words)
  print(f"num of words = {totalWords}")


  #cleaning process

  # # 1. new line delete
  strlst2=list(map(lambda x: x.strip(),str_lst))
  print(strlst2)

  #2. space delete
  strlst2 = list(map(lambda x : x.replace(" ",""),strlst2))
  print(strlst2)

  # 3. ch num
  ch_num=list(map(lambda x:len(x),strlst2))
  print(f"ch num = {sum(ch_num)}")

  with open("./sample_data/counter_of_string.txt","w") as f:
    f.write(f"num of lines = {len(str_lst)}\nnum of words = {totalWords}\nnum of ch = {sum(ch_num)}")
```

# Extra Practice Problem

Create a text file named emails.txt containing multiple email addresses, some in uppercase and some repeated.
Write a Python program to clean the data by converting all emails to lowercase, removing duplicates, sorting them alphabetically, and saving the cleaned list to a new file named unique_emails.txt.bold text

```python
with open("./sample_data/emails.txt","r") as f:
  emails=f.readlines()
  print(emails)

  #remove newline and lower
  emails= list(map(lambda x:x.strip().lower(),emails))
  print(emails)

  #remove dupli
  emails=set(emails)
  print(emails)
  #sort
  emails=sorted(emails)
  print(emails)

  with open("./sample_data/unique_emails.txt","w") as file:
    for email in emails:
      file.write(email + '\n')
```

Create a text file named names.txt containing a list of names.
Some names may have extra spaces, inconsistent capitalization (like “aLiCe” or “ bob ”), or appear more than once.
Write a Python program to clean the data by:
1.  Removing extra spaces.
2.  Converting all names so that only the first letter is capitalized.
3.  Removing duplicate names.
4. Sorting the names alphabetically.
Finally, save the cleaned list into a new file named clean_names.txt.


```python
with open("./sample_data/names.txt","r") as f:
  names=f.readlines()
  print(names)
  # remove spaces + normalize capitalization
  names=list(map(lambda x:x.strip().lower().capitalize(),names))
  print(names)
  #remove dupli
  names=set(names)
  print(names)
  names=sorted(names)
  print(names)

  with open("./sample_data/clean_names.txt","w") as file:
    for name in names:
      file.write(name + '\n')
```
