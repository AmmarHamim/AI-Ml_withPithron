# Different types of Arguments

```python
#positional arguments
def adder(a,b,c) :
  print(f"{a},{b},{c}")


adder(10,30,20)
```

```python
def adder(*args): #args takes inp as tuple
  print(args)

  for i in args:
    print(i)

adder(10,30,60,70)
```

```python
#keyword arguments
def adder(a,b,c,d):
  print(f"a={a},b={b},c={c},d={d}")

adder(10,100,30,69)
```

```python
def mul_keyargs(**kwargs): # takes inp as dict
  print(kwargs)

  for k,v in kwargs.items():
    print(f"{k}={v}")

mul_keyargs(a=20,b=40,x=50,y=69)
```

```python
def adder(a,b) :
  sum = a+b
  return sum

print(type(adder(10,20)))
```

# Iterator

```python
#iterating through a set
s = {10 , 20 , 30 , 40 , 50 }
it=iter(s)
print(next(it))
print(next(it))
print(next(it))
print(next(it))
print(next(it))
```

```python
#iterating through a dict(keys)
d={"name":"Hamim","age":20,"weight":70}
# print(type(d))

it=iter(d)
print(next(it))
print(next(it))
print(next(it))
```

```python
#iterating through a dict(values)
d={"name":"Hamim","age":20,"weight":70}
# print(type(d))

it=iter(d.values())
print(next(it))
print(next(it))
print(next(it))
```

```python
#iterating through a dict(key,values)
d={"name":"Hamim","age":20,"weight":70}
# print(type(d))

it=iter(d.items())
print(next(it))
print(next(it))
print(next(it))

print(type(it))
```

```python
it = iter(d.items())

while True:
    try:
        key, value = next(it)
        print(key, value)
    except StopIteration:
        break
```

# Generator

```python
lst = [ x for x in range(500) ] # dataset

def data_loader(chunk_size,lst):
    for i in range(0,len(lst),chunk_size):
        yield lst[i:i+chunk_size]


x = data_loader(5,lst)

print(next(x))
print(next(x))
print(next(x))
print(x)
print(type(x))
```

# Lemda,Map and Apply function

```python
import pandas as pd


df = pd.DataFrame({'temperature (F)': [32, 212, 98.6, 68]})
df

df['temperature (C)'] = df['temperature (F)'].apply(lambda x: (x - 32) * 5/9)
df
```

```python
import pandas as pd


df = pd.DataFrame({'A': [1, 2, 3, 4], 'B': [5, 6, 7, 8]})

df

df['A'] = df['A'].map(lambda x: x ** 0.5)


df
```

```python
import pandas as pd

df = pd.DataFrame({'total_spent': [120, 600, 450, 800]})

# Creating a 'High Value' feature on the fly
df['is_high_value'] = df['total_spent'].apply(lambda x: 1 if x > 500 else 0)

df
```
