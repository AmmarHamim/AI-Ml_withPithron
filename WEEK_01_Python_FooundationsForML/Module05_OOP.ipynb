# Class and Objects

```python
class Phone:
    category = "Electronics"

    def __init__(self, model, battery_capacity, camera_mp):
        self.model = model
        self.battery_capacity = battery_capacity
        self.camera_mp = camera_mp
        self.battery_percentage = 100

    def charge(self, hours):
        gained = hours * 10
        self.battery_percentage = min(100, self.battery_percentage + gained)
        print(f"{self.model} charged to {self.battery_percentage}%")

    def take_photo(self, count=1):
        required = count * 2
        if self.battery_percentage < required:
            print(f"{self.model}: Not enough battery")
            return
        self.battery_percentage -= required
        print(f"{self.model}: Captured {count} photo(s) 📸")

    def use(self, hours):
        consumption = hours * 5
        if self.battery_percentage < consumption:
            print(f"{self.model}: Battery too low!")
            return
        self.battery_percentage -= consumption
        print(f"{self.model}: Used for {hours} hour(s)")

    def status(self):
        print(f"""
📱 Model: {self.model}
🔋 Battery: {self.battery_percentage}%
📷 Camera: {self.camera_mp}MP
""")


# 🔷 Fixed class name
class CoolingMechanism:
    def __init__(self, cooling_type):
        self.cooling_type = cooling_type

    def cool(self):
        print(f"Cooling using {self.cooling_type} ❄️")


# 🔷 Fixed multiple inheritance class
class GamingPhone(Phone, CoolingMechanism):
    def __init__(self, model, battery, camera, cooling_type):
        Phone.__init__(self, model, battery, camera)
        CoolingMechanism.__init__(self, cooling_type)

    def play_game(self, hours):
        print(f"{self.model}: Gaming for {hours} hour(s) 🎮")
        self.use(hours * 2)
```

```python
apple = Phone("iPhone 17", 3000, 40)
motorola = Phone("Moto M17", 3500, 35)
print(apple.category)
apple.take_photo(22)
apple.use(4)
apple.status()
motorola.status()

g = GamingPhone("ROG Phone", 6000, 64, "Liquid Cooling")
print(g.battery_capacity)
print(g.cooling_type)
g.cool()
```

# Polymorphism

```python
class Camera:
  def __init__(self,name):
    self.name = name

  def capture(self):
    print(f"A photo is captured using {self.name}")
```

```python
class Phone(Camera):
  def __init__(self, name,resolution):
      super().__init__(name)
      self.resolution = resolution
#method overriding
  def capture(self):
    print(f"A photo is captured using {self.name} with resolution {self.resolution}")

class DSLR(Camera):
  def __init__(self, name,resolution):
      super().__init__(name)
      self.resolution = resolution

  def capture(self):
    print(f"A photo is captured using {self.name} with resolution {self.resolution}")

class Drone(Camera):
  def __init__(self, name,resolution):
      super().__init__(name)
      self.resolution = resolution

  def capture(self):
    print(f"A photo is captured using {self.name} with resolution {self.resolution}")


phone= Phone("ip13",30)
dslr= DSLR("zve10",40)
drone = Drone("air mavic",50)

phone.capture()
dslr.capture()
drone.capture()
```

# Encapsulation

```python
class User:
  def __init__ (self,name,password):
    self.name = name
    self.__password = password

  def get_password(self):
    return self.__password

  def set_password(self,password):
    self.__password = password
    print("password updated")
```

```python
u=User("Hamim","8450")
print(u.name)
# print(u.__password) #cant be accessed
print(u.get_password()) #can be accessed

password="1234" #not updated
print(u.get_password())

#using setter getter
u.set_password("1234")
print(u.get_password())


```

# Abstraction

```python
from abc import ABC,abstractmethod

class Animal(ABC):

  @abstractmethod
  def make_sound(self):
    pass


class Dog(Animal):
  def make_sound(self):
     print("Dog barks")

class Cat(Animal):
  def make_sound(self):
    print("Cat meows")
```

```python
# a=Animal() abstract class obj cant be created
# d=Dog() cant implement without implementing the abstracts methods
# c=Cat()

d=Dog()
c=Cat()

d.make_sound()
c.make_sound()
# a=Animal() # abstract class obj cant be created
```

```python
class Phone:
  def __init__(self,model,battery=100):
    self.model=model
    self.battery=battery

class Smartphone(Phone):
  def __init__ (self,model,os,battery=100):
    super().__init__(model,battery)
    self.os=os

class Gamingphone(Smartphone):
  def __init__(self,model,os,cooling_system,battery=100,):
    super().__init__(model,os,battery)
    self.cooling_system=cooling_system

  def start_game(self,name):
    print(f"Playing {name} on {self.model}")

```

```python
g=Gamingphone("ROG Phone","Android","Liquid Cooling")
print(g.model)
print(g.os)
print(g.battery)
g.start_game("Efootball")
```

```python
class MobilePhone():
  def __init__(self,model,battery):
    self.model=model
    self.__battery=battery

  def use(self):
    self.__battery-=10
  def increase(self):
    self.__battery+=5

  def get_battery(self):
    return self.__battery
```

```python
mp=MobilePhone("iphone",100)
# pritn(mp.battery)
print(mp.get_battery())
mp.use()
mp.use()
print(mp.get_battery())
mp.increase()
print(mp.get_battery())
```
