# nd array

```python
import numpy as np
#1d arr
arr1=np.array([2,3,6,7,5])
#2d arr
arr2=np.array([[1,2,3],[4,5,6]])
#2d arr
arr3=np.array([[1,2,3],
              [4,5,6],
              [5,5,5]])
#3d arr

arr4=np.array([
                #1st floor
               [
                   [1,2,3],
                   [4,5,6]
               ],
              #2nd floor
               [
                   [4,4,5],
                   [9,0,4]
               ]
])
```

```python
# dictionary

dc = { 'a' : 10 , 'b' : 20 , 'c' :30 }

keys = dc.keys()
values = dc.values()
items = dc.items()
```

```python
#keys to ndarray
arr = np.array(list(keys))
print(arr)
print(type(arr))
print(arr.dtype)
print(arr.shape)
print(arr.ndim)
print(arr.size)
```

```python
#values to ndarray
arr = np.array(list(values))
print(arr)
arr = arr.astype(np.int16)
print(type(arr))
print(arr.dtype)
print(arr.shape)
print(arr.ndim)
print(arr.size)
```

```python
#items to ndarray
arr = np.array(list(items))
print(arr)
print(type(arr))
print(arr.dtype)
print(arr.shape)
print(arr.ndim)
print(arr.size)
```

# array creation with range functions

```python
# np.arange(first , end , step ) -> similar to range function

arr=np.arange(1,13,2)

mat=arr.reshape(2,3)

print(arr)
print(mat)
print(type(arr))
print(arr.dtype)
print(arr.shape)
print(arr.ndim)
print(arr.size)
```

```python
# np.linspace(start , end(included) , koita)
# Creates numbers with equal spacing
# You specify how many values, not step

arr = np.linspace(0, 4 ,6)


print(arr)
print(type(arr))
print(arr.dtype)
print(arr.shape)
print(arr.ndim)
print(arr.size)
```

```python
# np.logspace(start , end(included) , koita )

# Creates numbers spaced logarithmically (exponential)
# Based on powers of 10 (default)

arr = np.logspace(0 , 4 ,6)
arr1 = np.logspace(0 , 4 ,6,base=2)

print(arr)
print(arr1)
```

## indexing and slicing

```python
#slicing
#1d = arr[start:end:step]

arr1=[2,3,6,7,8,99,22]

arr1_mod = arr1[1:4].copy()
print(arr1_mod)
print(arr1)


arr1[1] = 2
arr1_mod[2] = 200
print(arr1_mod)
print(arr1)
```

```python
#2d slicing
# arr[ row_start : row_end : step , column_start : column_end : step ]

import numpy as np

arr = np.array([[10, 20, 30],
                [40, 50, 60],
                [70, 80, 90]])

row0=arr[0:1,::]

col2=arr[::,2::]

slicemat=arr[1::,1::]

# print(arr[0,:])
print(row0)
print(col2)
print(slicemat)
```

1. You have a list of temperatures recorded in Celsius. Your task is to:
Create a NumPy array from the given list.
Convert all temperatures from Celsius to Fahrenheit. The formula is F = C* 1.8 + 32.
Print the new Fahrenheit array.


```python
import numpy as np
celsius = [0, 10, 20, 30, 40]

c=np.array(celsius)
# print(c)

f=c*1.8+32
print(f"{c} values in Farenheit is {f}")
```
